```python
import pandas as pd 
import numpy as np

import warnings
warnings.filterwarnings("ignore")

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
data=pd.read_csv("car_purchasing.csv",encoding='ISO-8859-1')
```

```python
dt=data.copy(deep=True)
```

```python
dt.columns
```

```python
dt.drop(["customer name","customer e-mail","country"],axis=1,inplace=True)
```

```python
dt.head()
```

```python
scaling_features=dt[['gender', 'age','annual Salary', 'credit card debt', 'net worth',]]
scaling_features
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
dt[['gender', 'age','annual Salary', 'credit card debt', 'net worth',]]=sc.fit_transform(scaling_features)
```

```python
dt
```

```python
x=dt.iloc[:,:-1]
print(x)
y=dt.iloc[:,-1]
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
ann=Sequential()

ann.add(Dense(units=6))
ann.add(Dense(units=8))
ann.add(Dense(units=6))
ann.add(Dense(units=1,activation="linear"))

ann.compile(optimizer="adam",loss="mean_squared_error",metrics=["mean_absolute_error"])

ann.fit(xtrain,ytrain,batch_size=10,epochs=300)

ypred=ann.predict(xtest)
```

```python
ypred=ann.predict(xtest)
```

```python
from sklearn.metrics import r2_score

r2_score(ytest,ypred)
```
