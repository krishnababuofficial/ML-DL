```python
data = raw_data.copy(deep=True)
```

```python
data.drop(data.loc[:, ['customer name', 'customer e-mail', 'country']], axis=1, inplace=True)
```

```python
data.info()
```

```python
scaling_features = data.loc[:, ['age', 'annual Salary', 'credit card debt', 'net worth']]
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
data.loc[:, ['age', 'annual Salary', 'credit card debt', 'net worth']] = scale.fit_transform(scaling_features)
```

```python
x = data.drop('car purchase amount', axis=1)
y = data['car purchase amount']
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=1, test_size=0.3)
```

```python
import tensorflow as tf

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
ann = Sequential()

ann.add(Dense(units=6))

ann.add(Dense(units=8))

ann.add(Dense(units=6))

ann.add(Dense(units=1, activation='linear'))

ann.compile(optimizer='adam', loss='mean_squared_error', metrics='mean_absolute_error')

ann.fit(xtrain, ytrain, batch_size=10, epochs=300)
```

```python
ypred = ann.predict(xtest)
```

```python
from sklearn.metrics import r2_score

r2_score(ytest, ypred)
```

```python
from sklearn.tree import DecisionTreeRegressor

rand = DecisionTreeRegressor()

rand.fit(xtrain, ytrain)

ypred1 = rand.predict(xtest)
```

```python
r2_score(ytest, ypred1)
```
