```python
sns.boxplot(data["Insulin"])
plt.show();

q3=data["Insulin"].quantile(0.75)
q1=data["Insulin"].quantile(0.25)
iqr=q3-q1
print(f'IQR: {iqr}\n')

upper_whsk=q3+1.5*iqr
lower_whsk=q1-1.5*iqr
print(f'UPPER_WHISKER:{upper_whsk}\nLOWER_WHISKER:{lower_whsk}\n')

upperwhisker_index=data[data["Insulin"]>upper_whsk].index
lowerwhisker_index=data[data["Insulin"]<lower_whsk].index
print(f'Upper Whisker Index:{upperwhisker_index}\nLower Whisker Index:{lowerwhisker_index}\n')

data.loc[upperwhisker_index,"Insulin"]=np.nan
data.fillna(upper_whsk,inplace=True)
data.loc[lowerwhisker_index,"Insulin"]=np.nan
data.fillna(lower_whsk,inplace=True)

sns.boxplot(data["Insulin"])
plt.show();
```

```python
scale=data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']]
scale
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']]=sc.fit_transform(scale)
```

```python
data
```

```python
x=data.drop("Outcome",axis=1)
print(x)
y=data["Outcome"]
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
ann=Sequential()
ann.add(Dense(units=200))
ann.add(Dense(units=200,activation="relu"))
ann.add(Dense(units=100))
ann.add(Dense(units=1,activation="sigmoid"))
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
ann.fit(xtrain,ytrain,batch_size=20,epochs=300,validation_data=(xtest,ytest))

ypred=ann.predict(xtest)
```

```python
ypred=ann.predict(xtest)
```

```python
ypred=ypred>0.5
```

```python
from sklearn.metrics import classification_report
```

```python
train=ann.evaluate(xtrain,ytrain)
test=ann.evaluate(xtest,ytest)

print(f'\ntrain: {train}\n test: {test}\n')
print(classification_report(ytest,ypred))
```
