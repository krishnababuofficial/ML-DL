```python
x=data.iloc[:,1:]
print(x)
y=data.iloc[:,0]
print(y)
```

```python
from keras.utils import np_utils

y=np_utils.to_categorical(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)

from sklearn.metrics import classification_report
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
ann=Sequential()
es=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=20)

ann.add(Dense(units=200))
ann.add(Dense(units=250))
ann.add(Dense(units=300))
ann.add(Dense(units=10,activation="softmax"))
ann.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
ann.fit(xtrain,ytrain,batch_size=40,epochs=100,validation_data=(xtest,ytest),callbacks=[es])
ypred=ann.predict(xtest)
```

```python
def mymod(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    
    train=model.score(xtrain,ytrain)
    test=model.score(xtest,ytest)
    
    print(f"Training Accuracy: {train}\nTesting Accuracy: {test}\n\n")
    print(classification_report(ytest,ypred))
    return model
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=mymod(KNeighborsClassifier())
```
